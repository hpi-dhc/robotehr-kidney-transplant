```python
import matplotlib.pyplot as plt

from robotehr.evaluation.plots import calibration_plot
from robotehr.models.predictor import Predictor
```

```python
predictor_uti = Predictor.load(id=20)
predictor_ovi = Predictor.load(id=19)
predictor_cmv = Predictor.load(id=17)
```

```python
plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = [r'\usepackage[cm]{sfmath}']
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'cm'
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[12,4])
fig.tight_layout(pad=3.0)

plt.yticks(rotation=0, fontsize=16)
plt.xticks(rotation=0, fontsize=16)

plt.rc('legend', fontsize=11)
calibration_plot(predictor_uti, 'LR', ax=ax[0], title="Urinary Tract Infection", show_brier=False)
calibration_plot(predictor_cmv, 'LR', ax=ax[1], title="Cytomegalovirus Infection", show_brier=False)
calibration_plot(predictor_ovi, 'LR', ax=ax[2], title="Other Viral Infection", show_brier=False)
ax[0].xaxis.label.set_size(18)
ax[0].yaxis.label.set_size(18)
ax[1].xaxis.label.set_size(18)
ax[1].yaxis.label.set_size(18)
ax[2].xaxis.label.set_size(18)
ax[2].set_ylabel('', fontsize=18)
ax[1].set_ylabel('', fontsize=18)
ax[0].title.set_size(20)
ax[1].title.set_size(20)
ax[2].title.set_size(20)

fig.savefig('/home/martet02/figures/calibration-plots-thesis.pdf', bbox_inches='tight')
```
