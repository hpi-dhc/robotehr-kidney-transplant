```python
plt.figure(figsize=[12,8])
tr = TrainingResult.load_by_config(
    training_configuration=get_training_configuration(
        pipeline_id=246,
        config={
            'threshold_numeric': 0.05, 
            'window_start_numeric': -361, 
            'target': 'urinary_tract_infection_onset_from_0_days_after_to_365_days_after'
        },
        response_type="object"
    ),
    algorithm='LogisticRegression',
    sampler='NoSampler'
)
clinical_usefulness_graph([tr], title="Urinary Tract Infection", label="Logistic Regression", ymax=0.35, ymin=-0.2)
```
