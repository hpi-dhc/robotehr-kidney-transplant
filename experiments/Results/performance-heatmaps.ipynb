# UTI

```python
# Implementation adopted from `plot_performance_heatmap` and tailored to NDM needs

plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = [r'\usepackage[cm]{sfmath}']
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'cm'
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[4,4])
fig.tight_layout(pad=3.0)

c = {
        'use case': 'AUROC Heatmap',
        'pipeline_id': 246,
        'algorithm': 'LogisticRegression',
        'target': 'urinary_tract_infection_onset_from_0_days_after_to_365_days_after',
}
metric="auc_roc_mean"
dimensions=['window_start_numeric', 'threshold_numeric']

df = get_training_results(c['pipeline_id'], metrics=[metric], response_type="pandas")
data = calculate_average_metric_score(df, c['algorithm'], metric, c['target'], dimensions)
vmin = math.floor(data.min().min() * 10) / 10.0
vmax = math.ceil(data.max().max() * 10) / 10.0
df = get_training_results(c['pipeline_id'], metrics=[metric], response_type="pandas")
data = calculate_average_metric_score(df, c['algorithm'], metric, c['target'], dimensions)

# replacing the days with months
data.columns = pd.Series(list(range(12,0,-1)))

sns.heatmap(
    data, vmin=vmin, cmap="YlGnBu",
    vmax=vmax, cbar=True, annot=False,
    linewidths=2, square=False,
    ax=ax
)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

ax.set_title(c['use case'], fontsize=18)
    
ax.set_ylabel('Fill Threshold', fontsize=14)

ax.set_xlabel('Beginning of OW', fontsize=14)


fig.savefig("/home/martet02/figures/performance-heatmap-uti.pdf", bbox_inches='tight')
```

## Algorithm + Sampler

```python
# Implementation adopted from `plot_performance_heatmap` and tailored to NDM needs

plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = [r'\usepackage[cm]{sfmath}']
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'cm'
plt.rcParams['font.size'] = 14

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[12,4])
fig.tight_layout(pad=3.0)

config = [
    {
        'use case': 'Urinary Tract Infection',
        'pipeline_id': 246,
        #'algorithm': 'LogisticRegression',
        'target': 'urinary_tract_infection_onset_from_0_days_after_to_365_days_after',
        
    },
    {
        'use case': 'Cytomegalovirus Disease',
        'pipeline_id': 231,
        #'algorithm': 'LogisticRegression',
        'target': 'cytomegaloviral_disease_onset_from_0_days_after_to_365_days_after',
        
    },
    {
        'use case': 'Other Viral Infection',
        'pipeline_id': 196,
        #'algorithm': 'LogisticRegression',
        'target': 'other_viral_infection_onset_from_0_days_after_to_365_days_after',
        
    },    
]
metric="auc_roc_mean"
dimensions=['algorithm', 'sampler']

for i in range(len(config)):
    c = config[i]
    df = get_training_results(c['pipeline_id'], metrics=[metric], response_type="pandas")
    data = df.groupby(dimensions).agg('mean').reset_index().pivot_table(index='algorithm', columns='sampler', values=metric)
    vmin = math.floor(data.min().min() * 10) / 10.0
    vmax = math.ceil(data.max().max() * 10) / 10.0
    
    # replacing the days with months
    data.columns = pd.Series(['bSMOTE', 'None', 'RandomO', 'RandomU', 'SMOTE'])
    data.index = pd.Series(['DT', 'GBDT', 'LR', 'RF'])
    
    sns.heatmap(
        data, vmin=vmin, cmap="YlGnBu",
        vmax=vmax, cbar=True, annot=False,
        linewidths=2, square=False,
        ax=ax[i]
    )
    bottom, top = ax[i].get_ylim()
    ax[i].set_ylim(bottom + 0.5, top - 0.5)

    ax[i].set_title(c['use case'], fontsize=20)
    
ax[0].set_ylabel('Algorithm', fontsize=16)
ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[0].set_yticklabels(ax[0].get_yticklabels(), rotation = 0)
ax[1].set_yticklabels(ax[1].get_yticklabels(), rotation = 0)
ax[2].set_yticklabels(ax[2].get_yticklabels(), rotation = 0)


ax[0].set_xlabel('Sampler', fontsize=16)
ax[1].set_xlabel('Sampler', fontsize=16)
ax[2].set_xlabel('Sampler', fontsize=16)

fig.savefig("/home/martet02/figures/performance-heatmaps-algorithms.pdf")
```
